# Prediction of H1N1 and Seasonal Flu Vaccinations
## 1. Introduction
Influenza(flu) is an illness that causes respiratory problems and can lead to hospitalisation or even death in some cases as a result of its complications. People over the age of 65, young children and people with pre-existing conditions are at a higher risk of flu-related [complications](https://www.cdc.gov/flu/about/index.html). An annual seasonal flu vaccine can help mitigate the effects of the flu by reducing its severity. The vaccine causes antibodies to develop after two weeks which provide protection against the flu. Everyone over 6 months of age is advised to take the vaccine according to the [Center for Disease Control and Prevention](https://www.cdc.gov/flu/prevent/keyfacts.htm).

The H1N1 pandemic began in 2009 and the influenza A virus first emerged in the United States and led to the deaths of approximately 151700-575400 people worldwide. The flu primarily affected young children and middle aged adults but surprisingly, people over 60 years old were less affected, presumably because they had antibodies against this virus, likely from exposure to an older H1N1 virus earlier in their lives. A vaccine for the flu became available to the public in October 2009 and consequently, the World Health Organisation (WHO) declared an end to the pandemic in August 2010. However, the virus continues to circulate as a seasonal [flu virus](https://www.cdc.gov/flu/pandemic-resources/2009-h1n1-pandemic.html).

### 1.1. Problem Statement
In 2019 a novel Coronavirus (COVID-19) was first identified in Wuhan, China, causing severe respiratory problems likened to pneumonia. There was (and still is) a lot of anxiety and reluctance associated with taking the vaccine due to its fast production among other reasons and many conspiracy theories came about as a result. Interestingly, the same reluctance was witnessed during the H1N1 pandemic period as reported by [Relias Media](https://www.reliasmedia.com/articles/114699-h1n1-vaccine-fears-may-fuel-pandemic-undercut-hospital-public-health-response#:~:text=Public%20health%20officials%20are%20concerned%20about%20the%20perceived%20reluctance%20of%20millions%20of%20people%20to%20be%20immunized%20with%20a%20fast%2Dtracked%20vaccine%20for%20what%20many%20dismiss%20as%20a%20mild%20infection) as well as in this article by [Seale, H., Heywood, A.E., McLaws, ML. et al.](https://bmcinfectdis.biomedcentral.com/articles/10.1186/1471-2334-10-99) on the BMC website.

> Vaccine hesitancy is the unwillingness to receive vaccines due to fears of the vaccine's effects on one's body and is one of the biggest threats to global health. People's views on vaccination are influenced by the personal and social circumstances and they can change over time. Unfortunately, the effects of vaccine hesitancy are more pronounced during pandemics and pose huge risks to the wider community as denoted in this paper by [Wiysonge CS, Ndwandwe D. et al.](https://pubmed.ncbi.nlm.nih.gov/33684019/).

A strong correlation has also been detected between people that did not take the H1N1 vaccine and their willingness to take the COVID-19 vaccine. This paper by [Nair P, Wales DP.](https://www.ncbi.nlm.nih.gov/pmc/articles/PMC8888253/#:~:text=Of%20those%20who%20did%20not%20receive%20the%202009%20H1N1%20vaccine%2C%2092.9%25%20of%20them%20disagreed%20with%20getting%20the%20COVID%2D19%20vaccine) reveals that 92.9% of people that failed to take the H1N1 vaccine would not want to take the COVID-19 vaccine. 

It is up to scientists and public health practitioners to identify the reasons behind this and work towards mitigating them. The first step could be using information people's socioeconomic status and demographic backgrounds, beliefs as well as their opinions on vaccine effectiveness and risks in order to identify potential reasons for vaccine hesitancy thus working on public awareness campaigns and creating new public health policies thus reducing the risks involved with vaccine hesitancy for future pandemics.

## 6. Data Preprocessing
This section aims to prepare the data for modelling. It addresses missing data, wrong data types, invalid datatypes, scaling, encoding any categorical variables,dropping highly correlated columns and any feature engineering needed. The data is first split into training and test data to prevent data leakage. The target variables are meant to be `h1n1_vaccine` and `seasonal_vaccine`. The two columns are both droped from X. This section focuses on `seasonal_vaccine` as the target variable. 

```python
# separating target variables from the features
X = data.drop(["seasonal_vaccine","h1n1_vaccine"],axis=1)
y = data.seasonal_vaccine
```

```python
# splitting data into test and train and setting a random seed for reproducibility
random_seed = 1234

X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=random_seed,stratify=y)
```

#### 6.1. Checking for Multicollinearity
Columns that are highly correlated will lead to multicollinearity issues which can affect model performance as there are two columns providing similar information. One of the columns would need to be dropped. A dataframe is created with correlated pairs and their correlation coefficent. Anything above 80% should be have one column dropped

```python
# creating dataframe of correlated pairs and sorting in descending order
df_corr = X_train.corr().abs().stack().reset_index().sort_values(0,ascending = False)
df_corr["pairs"] = list(zip(df_corr.level_0,df_corr.level_1))
df_corr.set_index(["pairs"],inplace = True)
df_corr.drop(columns = ["level_1","level_0"],inplace=True)
df_corr.columns = ["cc"]
df_corr.drop_duplicates(inplace = True) 
```

```python
# previewing top ten correlated pairs
df_corr[:10]
```

The highest correlation is 59%. As a result, no columns will be dropped.

### 6.2. Missing values
Missing values are to be dealt with in a number of ways:
* replacing with mode or the string 'missing'
* dropping rows

#### 6.2.1 Replacing
The SimpleImputer class is used to impute columns with missing values with their modes and with string "missing".

```python
# previewing columns with missing data
missing_train = X_train.isna().sum().to_frame().sort_values(0,ascending = False)
missing_train
```

```python
# instantiating a SimpleImputer to fill missing data with their modes
imputer_fill = SimpleImputer(strategy = "most_frequent",missing_values = np.nan)
cols_to_fill = ["income_poverty","doctor_recc_h1n1","doctor_recc_seasonal","education","health_insurance","health_worker"]

# fitting and transforming on the train data then reassigning 
X_train[cols_to_fill] = imputer_fill.fit_transform(X_train[cols_to_fill])

# transforming test data
X_test[cols_to_fill] = imputer_fill.transform(X_test[cols_to_fill])

```

##### 6.2.2. Dropping rows

At this point rows can be dropped because the amount of data that will be lost is very small

```python
# dropping rows with null values
X_train.dropna(inplace=True)
X_test.dropna(inplace=True)
```

It also makes sense to remove the dropped rows from y_test and y_train for the models to work.

```python
# identifying indices to drop
set1 = set(y_test.index) 
set2 = set(X_test.index)
set3 = set(y_train.index)
set4 = set(X_train.index)

index_to_drop_test = list(set1 - set2)
index_to_drop_train = list(set3-set4)
```

### 6.3. Encoding Categorical Values
Since most machine learning models are unable to work with categorical values, there is need to represent these values in numerical format. This section addresses this using the pandas `get_dummies` method. There are also three columns that have been encoded with random strings. They are to be replaced with numbers for readability then one hot encoded.

#### 6.3.1. Replacing column values
This can be achieved with the OrdinalEncorder class which will result in a column of integers from 0 to n-1 categories.

```python
# assessing one of the columns with anonymised data
X_train.employment_industry.value_counts()
```

```python
# columns to replace
to_replace = ['employment_industry','employment_occupation','hhs_geo_region']

# instantiating the class
ord_en = OrdinalEncoder()

# fitting and transforming on training data and transforming on test data
X_train[to_replace] = ord_en.fit_transform(X_train[to_replace])
X_test[to_replace] = ord_en.transform(X_test[to_replace])
```

```python
# confirming the change
X_train.employment_industry.value_counts()
```

##### 6.3.2 One Hot Encoding
All the columns with the float and integer data types are categorical in nature thus necessitating their change into the category data type in preparation for one hot encoding

```python
# assigning columns with float datatype to a variable
float_cols = X_train.select_dtypes(["float","int"]).columns

# for loop to convert datatype to category
for col in float_cols:
    X_train[col] = X_train[col].astype("category")
    X_test[col] = X_test[col].astype("category")

print(X_train.dtypes.value_counts(),"\n")
print(X_test.dtypes.value_counts()  )  
```

```python
# one hot encoding the columns
X_train_one_hot = pd.get_dummies(X_train,drop_first=True)
X_test_one_hot = pd.get_dummies(X_test,drop_first=True)
```

At this point the dataset is simply a bunch of 1s and 0s. There is therefore no need for scaling. The data is ready for modelling.

## 7. Modelling
This section look at the iterative modelling process in a bid to arrive at one with the highest accuracy on the test data. A baseline model is built first and its accuracy evaluated. Cross validation is also done so as to have a more accurate estimate of model performance.......
### 7.1. Baseline Model
A logistic regression model is used as the baseline model as we are dealing with a classification problem. A random state is assigned for reproducibility.

```python
# defining function that fits a model and returns accuracy
def evaluation(model,X_train,y_train,X_test,y_test):
    
    # fitting the model
    model.fit(X_train,y_train)

    # generating predictions for train and test data
    y_train_pred = model.predict(X_train)
    y_test_pred = model.predict(X_test)

    #performing cross validation
    cross_val = cross_val_score(model,X_train,y_train)

    #printing accuracy
    print(f"Train Accuracy: {round(accuracy_score(y_train,y_train_pred)*100,5)}%")
    print(f"Cross Validation Accuracy: {round(cross_val.mean()*100,5)}%")
    print(f"Test Accuracy: {round(accuracy_score(y_test,y_test_pred)*100,5)}%")
```

```python
# instantiating LogisticRegression class
baseline_model = LogisticRegression(random_state=random_seed)

#evaluating model
print("Logistic Regression Baseline Model Scores")
evaluation(baseline_model,X_train_one_hot,y_train,X_test_one_hot,y_test)
```

This model is unfortunately not able to converge so a solver is defined as sag which is recommended for [large datasets](https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LogisticRegression.html#:~:text=whereas%20%E2%80%98sag%E2%80%99%20and%20%E2%80%98saga%E2%80%99%20are%20faster%20for%20large%20ones) according to the documentation.

```python
# instantiating model with solver defined
baseline_new = LogisticRegression(random_state=random_seed,solver = "sag")

#evaluating model
print("Logistic Regression New Baseline Model Scores")
evaluation(baseline_new,X_train_one_hot,y_train,X_test_one_hot,y_test)
```

The validation score is very close to the accuracy for training data which is a good sign. The difference in the accuracy for train and test data is small, which is good for the baseline model, it is an indication that there is no overfitting.

##### 7.2.2 Tuning RandomForest Hyperparameters
GridSearchCV is used to find the best combination of hyperparameters and the following are explored for the stated reasons:

* max_leaf_nodes - this hyperparameter will restrict the growth of a tree by specifiying the maximum number of terminal/leaf nodes allowed after which no more splitting can take place.
* min_samples_split - this hyperparameter specifies the minimum number of observations in a given node before splitting. A higher number means there is less of a chance of overfitting.
* n_estimators - this specifying the number of trees that shoud be built(planted?). The more the number of trees, the more complex the model is. It would help to find a sweet spot.

```python
params = [{"max_leaf_nodes":[100,150,200],
"min_samples_split":range(2,6),
"n_estimators":[100,150,200]}]

grid = GridSearchCV(rf,param_grid=params,cv=7)

grid.fit(X_train_rfe,y_train)

print("The best parameters are \n",grid.best_params_)
```

```python
rf_tuned = RandomForestClassifier(random_state=random_seed,max_leaf_nodes=200,min_samples_split=2,n_estimators=150)

print("Random Forest Model with tuned hyperparameters:")
evaluation(rf_tuned,X_train_rfe,y_train,X_test_rfe,y_test)
```

```python
# previewing classification report 
pred_test_tuned_rf = rf_tuned.predict(X_test_rfe)
print(classification_report(y_test,pred_test_tuned_rf))
```

There is a marked improvement in the accuracy, from 76.9% to 77.4%. The model is also no longer overfitting. However, the logistic regression model still has a higher accuracy of 78.04%.
